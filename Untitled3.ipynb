```python
def build_features_targets(customer_df, T):
    # ensure sorted by time (important for recency)
    customer_df = customer_df.sort_values('invoicedate1')

    # split past and future
    past = customer_df[customer_df['invoicedate1'] < T]
    future = customer_df[customer_df['invoicedate1'] >= T]

    features = {}

    # past spend & order windows
    for d in [30, 60, 90, 180]:
        past_window = past[
            past['invoicedate1'] >= (T - pd.Timedelta(days=d))
        ]

        features[f'spend_past_{d}d'] = past_window['TotalAmount'].sum()
        features[f'orders_past_{d}d'] = past_window['InvoiceNo'].nunique()

    # recency (days since last purchase before T)
    if not past.empty:
        features['recency_days'] = (T - past['invoicedate1'].iloc[-1]).days
    else:
        features['recency_days'] = 999  # cold-start customer

    targets = {}

    # future spend targets
    for d in [30, 60, 90, 180]:
        future_window = future[
            future['invoicedate1'] < (T + pd.Timedelta(days=d))
        ]

        targets[f'target_spend_{d}d'] = future_window['TotalAmount'].sum()

    return features, targets
```

```python
def generate_cutoffs(
    customer_df,
    start_date="2011-04-01",
    prediction_horizons=[30, 60, 90, 180],
    step_days=30
):
    start_date = pd.to_datetime(start_date)

    max_horizon = max(prediction_horizons)

    # last possible cutoff so that future target window exists
    last_date = customer_df['invoicedate1'].max() - pd.Timedelta(days=max_horizon)

    # if customer doesn't have data reaching the horizon, return empty
    if start_date > last_date:
        return pd.DatetimeIndex([])

    return pd.date_range(
        start=start_date,
        end=last_date,
        freq=f'{step_days}D'
    )
```

```python
rows = []

# IMPORTANT: sort once globally (performance + correctness)
df_no_nulls = df_no_nulls.sort_values(['CustomerID', 'invoicedate1'])

for customer_id, customer_df in df_no_nulls.groupby('CustomerID'):

    # generate valid cutoff dates for this customer
    cutoffs = generate_cutoffs(customer_df)

    # skip customers with no valid cutoffs
    if len(cutoffs) == 0:
        continue

    for T in cutoffs:
        features, targets = build_features_targets(customer_df, T)

        row = {
            'CustomerID': customer_id,
            'cutoff_date': T
        }

        row.update(features)
        row.update(targets)

        rows.append(row)

# create final training dataframe
training_df = pd.DataFrame(rows)
```

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from xgboost import XGBRegressor
```

```python
FEATURES = [
    'spend_past_30d','orders_past_30d',
    'spend_past_60d','orders_past_60d',
    'spend_past_90d','orders_past_90d',
    'spend_past_180d','orders_past_180d',
    'recency_days'
]
```

```python
training_df = training_df.sort_values('cutoff_date')

X = training_df[FEATURES]

split_idx = int(len(training_df) * 0.8)

X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
```

```python
def evaluate(y_true, y_pred):
    return {
        "MAE": mean_absolute_error(y_true, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),
        "R2": r2_score(y_true, y_pred)
    }
```

```python
models = {
    "LinearRegression": LinearRegression(),

    "RandomForest": RandomForestRegressor(
        n_estimators=300,
        max_depth=12,
        min_samples_leaf=20,
        random_state=42,
        n_jobs=-1
    ),

    "XGBoost": XGBRegressor(
        n_estimators=400,
        max_depth=6,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="reg:squarederror",
        random_state=42
    )
}
```

```python
results = []

TARGETS = [
    'target_spend_30d',
    'target_spend_60d',
    'target_spend_90d',
    'target_spend_180d'
]

trained_models = {}

for target in TARGETS:
    y = training_df[target]
    y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

    for model_name, model in models.items():
        model.fit(X_train, y_train)

        preds = model.predict(X_test)
        preds = np.maximum(preds, 0)

        metrics = evaluate(y_test, preds)

        horizon = target.replace("target_spend_", "")

        results.append({
            "Horizon": horizon,
            "Model": model_name,
            **metrics
        })

        #  STORE THE TRAINED MODEL
        trained_models[(horizon, model_name)] = model
```

```python
# create results dataframe
results_df = pd.DataFrame(results)

# select best model per horizon using MAE
best_models_mae = (
    results_df
    .sort_values("MAE")      # lowest MAE is best
    .groupby("Horizon")
    .first()
    .reset_index()
)

best_models_mae
```

```python
import joblib
import os

os.makedirs("saved_models", exist_ok=True)

for _, row in best_models_mae.iterrows():
    horizon = row["Horizon"]     # '30d', '60d', etc.
    model_name = row["Model"]

    best_model = trained_models[(horizon, model_name)]

    joblib.dump(
        best_model,
        f"saved_models/{model_name}_{horizon}.joblib"
    )
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Define the target columns (log-transformed)
target_cols = [
    'log_target_spend_30d',
    'log_target_spend_60d',
    'log_target_spend_90d',
    'log_target_spend_180d'
]

# Dictionary to store trained models
linear_models = {}

for target_col in target_cols:
    # Define features (X) and current target (y)
    features = [c for c in training_df.columns if c.startswith('spend_past') or c.startswith('orders_past') or c == 'recency_days']
    X = training_df[features]
    y = training_df[target_col]

    # Split data into training and testing sets for the current target
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, shuffle=False, test_size=0.2
    )

    # Initialize and train the Linear Regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Store the trained model in the dictionary
    linear_models[target_col] = {
        'model': model,
        'X_test': X_test,
        'y_test': y_test
    }

    print(f"Linear Regression model for {target_col} trained successfully.")

print("All Linear Regression models trained and stored in 'linear_models' dictionary.")
```

## Summary:

### Q&A
1.  **Why is it important to transform skewed target variables for machine learning models, and why is `np.log1p` a suitable method?**
    Skewed target variables can negatively impact machine learning models by violating assumptions (e.g., normality of residuals, homoscedasticity), leading to poor performance, increased outlier sensitivity, and unstable gradients. Logarithmic transformations, specifically `np.log1p(x)`, are suitable because they reduce skewness by compressing larger values and expanding smaller ones, bringing the distribution closer to normal. `np.log1p` is particularly useful as it handles zero values gracefully (returning 0 for an input of 0), unlike standard `np.log`, which would return undefined or problematic values. This makes it robust for datasets containing zero-valued targets, such as customer spend where zero spend is common.

*   The distributions of the newly transformed `log_` target variables were visualized using histograms, confirming the effect of the transformation in reducing skewness.
*   The dataset was split into features (`X`) and a target (`y`), with `X` containing 9 features (e.g., `spend_past`, `orders_past`, `recency_days`) and `y` specifically chosen as `log_target_spend_30d`. `X` had a shape of (5498, 9) and `y` a shape of (5498,).

*   A `RandomForestRegressor` model was trained with `n_estimators=200` on the `X_train` and `y_train` datasets.

### What is Multicollinearity?
Multicollinearity is a phenomenon in multiple regression where two or more predictor variables (features) in a model are highly correlated with each other. This means that one predictor variable can be linearly predicted from the others with a substantial degree of accuracy. While perfect multicollinearity (where one variable is a perfect linear combination of others) is rare and often detectable by software, high (near-perfect) multicollinearity is more common and can cause significant problems.

### Why is Checking for Multicollinearity Important?
Identifying and addressing multicollinearity is crucial in machine learning and statistical modeling for several reasons:

1.  **Reduced Model Interpretability**: When predictor variables are highly correlated, it becomes difficult to isolate the individual effect of each variable on the response variable. The coefficients assigned to these correlated variables can become unstable and highly sensitive to small changes in the data, making it hard to interpret their true relationship with the target.

2.  **Unstable Coefficient Estimates**: Multicollinearity inflates the variance of the regression coefficients, leading to larger standard errors. This makes the coefficients less reliable and less precise, meaning their estimated values can fluctuate wildly with different samples of data, or even with minor changes to the existing dataset.

3.  **Difficulty in Feature Selection**: It can mislead feature selection processes. A variable might appear statistically insignificant due to multicollinearity even if it is fundamentally related to the response variable, simply because its effect is already 'captured' by another correlated variable.

4.  **Impact on Prediction Accuracy (in some cases)**: While multicollinearity doesn't always drastically reduce the model's overall predictive power, it can lead to less robust predictions, especially when extrapolating to new data points that might have different correlation structures among the features. In some models, it can contribute to overfitting or less generalizable results.

5.  **Issues with Causal Inference**: If the goal is to understand the causal impact of individual features, multicollinearity makes it nearly impossible to draw accurate conclusions about the unique contribution of each correlated predictor.

### What are Outliers?
Outliers are data points that significantly deviate from other observations in a dataset. They are extreme values that lie an abnormal distance from other values. Outliers can arise due to various reasons, including measurement errors, data entry errors, natural variations in the data, or unique events.

### Why is it Important to Inspect and Understand Outliers?
Inspecting and understanding outliers is crucial in data analysis and machine learning for several reasons:

1.  **Data Quality and Integrity**: Outliers can often be indicators of errors in data collection, measurement, or entry. Identifying and understanding them can help in improving data quality.
2.  **Accurate Statistical Inference**: Outliers can disproportionately influence statistical measures such as the mean and standard deviation. Their presence can distort data distributions and lead to incorrect conclusions about the underlying population.
3.  **Model Performance**: Outliers can significantly impact the training and performance of machine learning models. Understanding their presence allows for informed decisions on how to handle them.
4.  **Discovery of Anomalies/Novelties**: Sometimes, outliers are not errors but represent rare, but important, events or behaviors (e.g., fraudulent transactions, system failures). Their detection can lead to valuable insights.

### Potential Negative Impacts of Outliers on Machine Learning Models:

The impact of outliers varies depending on the type of machine learning model:

*   **Regression Models (e.g., Linear Regression, Ridge, Lasso)**:
    *   **Coefficient Bias**: Outliers can pull the regression line towards themselves, leading to biased and inaccurate coefficient estimates. This results in a model that poorly generalizes to unseen data.
    *   **Increased Error Variance**: They can inflate the variance of the residuals, making the model's predictions less reliable.

*   **Tree-Based Models (e.g., Decision Trees, Random Forests, Gradient Boosting)**:
    *   **Less Sensitive, but still affected**: Tree-based models are generally more robust to outliers because they make decisions based on splits rather than distances or means. However, extremely influential outliers can still cause trees to create splits specifically to isolate these points, leading to complex and potentially overfitted models, especially with deep trees.
    *   **Impurity Measures**: Outliers can affect impurity measures (like Gini impurity or entropy) used for splitting, potentially leading to suboptimal tree structures.

*   **Clustering Models (e.g., K-Means)**:
    *   **Centroid Shift**: In algorithms like K-Means, outliers can heavily influence the position of cluster centroids, leading to incorrect cluster assignments and distorted cluster shapes.
    *   **Formation of Small Clusters**: Outliers might form their own small, isolated clusters, or pull other legitimate data points into their cluster.

*   **Distance-Based Models (e.g., K-Nearest Neighbors, Support Vector Machines)**:
    *   **KNN**: Outliers can distort the local neighborhood used for classification or regression, leading to misclassifications or inaccurate predictions for neighboring points.
    *   **SVM**: Outliers can influence the placement of the decision boundary, especially in hard-margin SVMs, potentially reducing the model's generalization capability. Soft-margin SVMs are more robust but can still be affected.

*   **Neural Networks**:
    *   **Training Instability**: Outliers can lead to larger gradients during backpropagation, causing the network's weights to update aggressively and making the training process unstable or slower to converge.
    *   **Biased Weights**: The network might learn weights that are overly influenced by the extreme values, leading to poor performance on typical data points.
